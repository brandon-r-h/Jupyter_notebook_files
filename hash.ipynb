```python
# While Hash Tables tradition use key value pairs
# For this assignment you may assume the value may be hashed directly. 

# In practice, the only changes to support key/value pairs would be in __getKey()

# Be sure to implement rehashing

# Be sure to use some other data structure internally

class HashTable:

    # set up the hash table, with no values in it
    def __init__(self):
        self.size = 0
        self.table = [None]*self.size

    # given some input, figure out its hash values
    # * You may use python hash() in this method
    def __getKey(self, x):
        return x%len(self.table)
        
    # rehash the table if it is not of appropriate size
    # * Rehashing is required for this assignment, though this is private
    # * I reserve the write to change the parameters of random list generation in autograder to force rehashing
    def __rehash(self):
        l = []
        for x in self.table:
            if x!=None:
                for y in x:
                    l.append(y)
        if self.size>len(self.table)//2:
             self.table = [None]*(len(self.table)*2)
        if self.size<len(self.table)//4:
            self.table = [None]*(len(self.table)/2)
        
        for x in l:
            self.insert(x)
        
            
    # insert x into the hash table
    # * Use some data structure to deal with collisions, such as:
    #   - Python List
    #   - Linked List
    #   - BST
    #   - Python Set
    #   - Python Dictionary
    #   - Stack
    #   - Queue
    def insert(self, x):
        if len(self.table)==0:
            self.table = [None,None,None,None]
            loc = self.__getKey(x)
            self.table[loc] = [x]
            self.size+=1
        else:
            loc = self.__getKey(x)
            if self.table[loc] ==None:
                self.table[loc] = [x]
            else:
                
                self.table[loc].append(x)
                self.size+=1
                if self.size>len(self.table)//2:
                    self.__rehash()
        
        
        
    # remove x from the hash table
    def remove(self, x):
        loc = self.__getKey(x)
        if self.table[loc]==None:
            return 
        for i in range(len(self.table[loc])):
            if self.table[loc][i]==x:
                self.table[loc].pop(i)
                self.size-= 1
                if self.size<len(self.table)//4:
                    self.__rehash()
            return
        return
        
                
                
    # check if x is in the hash table
    def contains(self, x):
        if len(self.table)==0:
            return False
        loc = self.__getKey(x)
        if self.table[loc] == None:
            return False
        
        if x in self.table[loc]:
            return True
        return False
```
