## Problem 1
Find the set of cars that has an MSRP of less than $45,000 and is Converible

## Answer

```python

cars[(cars['MSRP'] <45000)&(cars['Vehicle Style']=='Convertible')]
```

## Problem 2
Find the set of cars that are made by Toyota, have 4 doors, have more than 25 Highway MPG made after 2016

## Answer

```python

cars[(cars['Make'] =='Toyota')&(cars['Number of Doors']==4)&(cars['highway MPG']>25)&(cars['Year']>2016)]
```

## Problem 3
6 Cylinder Ford with less than 12 City MPG
## Answer

```python
 cars[(cars['Make'] =='Ford')&(cars['Engine Cylinders']==6)&(cars['city mpg']<12)]
```

## Problem 4

Find the car with the highest Highway MPG under $30,000 with a manuel transmission
## Answer

```python
cars[(cars['MSRP'] <30000)&(cars['Transmission Type']=='MANUAL')].sort_values(by=['highway MPG'],ascending = False).head(1)

```

## Problem 5

Find the car with the highest HP in 2015 by Ford.
## Answer

```python
cars[(cars['Year']==2015)&(cars['Make']=='Ford')].sort_values(by=['Engine HP'],ascending = False).head(1)
```

## Problem 6
What's the best car with the highest highway MPG with over 300 HP?
## Answer

```python
cars[cars['Engine HP']>300].sort_values(by=['highway MPG'],ascending = False).head(1)
```

## Problem 9
What is the lowest priced car that has 4 wheel drive with premium unleaded fuel type that is a luxury car with over 300 HP?
## Answer

```python
cars[(cars['Driven_Wheels']=='four wheel drive')&(cars['Engine Fuel Type']=='premium unleaded (required)')|(cars['Engine Fuel Type']=='premium unleaded (recommended)')&(cars['Engine HP']>300)&(cars['Market Category']=='Luxury')].sort_values(by=['MSRP'],ascending = True).head(1)
```
