```python

class HashTable:

    # set up the hash table, with no values in it
    def __init__(self):
        self.size = 0
        self.table = [None]*self.size

    # given some input, figure out its hash values
    # * You may use python hash() in this method
    def __getKey(self, x):
        return len(x)%len(self.table)
        
    # rehash the table if it is not of appropriate size
    # * Rehashing is required for this assignment, though this is private
    # * I reserve the write to change the parameters of random list generation in autograder to force rehashing
    def __rehash(self):
        l = []
        for x in self.table:
            if x!=None:
                for y in x:
                    l.append(y)
        if self.size>len(self.table)//2:
             self.table = [None]*(len(self.table)*2)
        if self.size<len(self.table)//4:
            self.table = [None]*(len(self.table)/2)
        
        for x in l:
            self.insert(x)
        
            
    
    def insert(self, x):
        if len(self.table)==0:
            self.table = [None,None,None,None]
            loc = self.__getKey(x[0])
            self.table[loc] = [x]
            self.table[loc][0][2] = [self.table[loc][0][2]]
            self.size+=1
            
            
        else:
            loc = self.__getKey(x[0])
            if self.table[loc] ==None:
                self.table[loc] = [x]
                self.table[loc][0][2] = [self.table[loc][0][2]]
            else:
                if self.contains(x[0])==True:
                    for i in range(0,len(self.table[loc])):
                        if self.table[loc][i][0]==x[0]:
                            self.table[loc][i][2].append(x[2])
                if self.contains(x[0])==False:
                    self.table[loc].append(x)
                    self.table[loc][-1][2] = [self.table[loc][-1][2]]
                    
                        
                if self.size>len(self.table)//2:
                    self.__rehash()
    def returnVal(self,x):
        if self.contains(x):
            loc = self.__getKey(x)
            for y in self.table[loc]:
                if y[0]==x:
                    return y
        else:
            return False
        
   
                
        
    # remove x from the hash table
    def remove(self, x):
        loc = self.__getKey(x)
        if self.table[loc]==None:
            return 
        for i in range(len(self.table[loc])):
            if self.table[loc][i]==x:
                self.table[loc].pop(i)
                self.size-= 1
                if self.size<len(self.table)//4:
                    self.__rehash()
            return
        return
        
                
                
    # check if x is in the hash table
    def contains(self, x):
        if len(self.table)==0:
            return False
        loc = self.__getKey(x)
        if self.table[loc] == None:
            return False
        for i in range(0,len(self.table[loc])):
            if self.table[loc][i][0]==x:
                return True
        return False

class Donation:
    
    def __init__(self,file):
        self.data = pd.read_csv(file, sep = ',')
        
        
        self.names = self.data['contributor_name'].tolist()
        self.donees = self.data['committee_name'].tolist()
        self.donationAmount = self.data['contribution_receipt_amount'].tolist()
        for i in range(0,len(self.donationAmount)):
            self.donationAmount[i] = int(self.donationAmount[i])
        
        self.donation = []
        self.l = HashTable()
        for i in range(0,len(self.names)):
            self.donation.append([self.names[i],self.donees[i],self.donationAmount[i]])
        for x in self.donation:
            self.l.insert(x)
            
    def donations(self):
        return self.l.table
    def contains(self,x):
        return self.l.contains(x)
    def insert(self,x):
        self.l.insert(x)
    def allDon(self,x):
        z = self.l.returnVal(x)
        amount = ""
        totamount = 0
        if z != False:
            for money in z[2]:
                amount = amount + "$"+str(money)+", "
                totamount+=money
                
            return x+":"+amount+" total amount:$"+str(totamount)
   
```
