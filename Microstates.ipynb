```python
import math
import matplotlib.pyplot as plt
```

```python
class Microstates:
    
    def __init__(self,quanta,object1,object2):
        self.quanta = quanta
        self.objects = [object1,object2]
    
    def microstates(self,quan,atom):
        return int(math.factorial(quan+self.objects[atom]-1)/(math.factorial(quan)*math.factorial(self.objects[atom]-1)))
    
    def allMS(self):
        x = []
        y = []
        q = []
        
        for i in range(0,self.quanta+1):
            x.append(self.microstates(i,0))
        for i in range(self.quanta,-1,-1):
            y.append(self.microstates(i,1))
        for i in range(0,self.quanta+1):
            q.append(x[i]*y[i])
        return q
            
    
    #def totalMs(self):
        #return math.factorial(self.quanta+sum(self.objects[1:])-1)/(math.factorial(self.quanta)*math.factorial(sum(self.objects[1:])-1))
    def plotMs(self):
        fig = plt.figure()
        ax = fig.add_axes([0,0,1,1])
        a = [i for i in range(0,self.quanta+1)]
        ax.bar(a,self.allMS())
        plt.show()
        
```

```python
x = Microstates(100,50,25)
x.plotMs()
```

```python
x = Microstates(120,100,250)
x.plotMs()
```
